```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_csv("data/token_df.csv")
```

```python
import scipy.stats as stats

stats.kstest(data_df["salary_standardized"], stats.norm.cdf, args=(data_df["salary_standardized"].mean(), data_df["salary_standardized"].std()))
```

```python
import seaborn as sns

sns.histplot(data_df["salary_standardized"],bins=30,kde=True).set(title="Salary Standardized for Data Jobs")

```

```python
sns.boxplot(data = data_df["salary_standardized"],showfliers = False).set(title="Standardized Salary for Data Jobs")
```

```python
#STANDARD SALARY VS REMOTE

remote=data_df[data_df["work_from_home"]==True]
sns.histplot(data=remote,x="salary_standardized",bins=30,kde=True)
```

```python
stats.kstest(remote["salary_standardized"], stats.norm.cdf, args=(remote["salary_standardized"].mean(),remote["salary_standardized"].std()))
```

```python
#STANDARD SALARY VS NON-REMOTE
non_remote=data_df[data_df["work_from_home"]==False]
sns.histplot(data=non_remote,x="salary_standardized",bins=30,kde=True).set(title="Salary for Non-Remote Data Jobs")
```

```python
stats.kstest(non_remote["salary_standardized"], stats.norm.cdf, args=(non_remote["salary_standardized"].mean(),non_remote["salary_standardized"].std()))
```

```python
#REMOTE vs NON-REMOTE Salaries
sns.boxplot(data = data_df, x = "work_from_home", y = "salary_standardized",showfliers = False).set(title="Salary by Non_Remote vs Remote")
```

```python
#RANK TOP 5 DATA SKILLS
skills=data_df.loc[:,"airflow":].sum().sort_values(ascending=False).head(20)

sns.barplot(x = skills.values, y = skills.index).set(title="Top 20 Data Skills")
```

```python
#TOP 5 CITIES
top_locations=data_df["location"].value_counts().head(20)

sns.barplot(x = top_locations.values, y = top_locations.index).set(title="Top 20 Data Job Locations")
```
