```python
import pandas as pd

data_df=pd.read_csv("/Users/Jackie/Desktop/GH_Clones/Big-Data-AWS-ETL/data/token_df.csv")
data_df.head()
```

```python
import scipy.stats as stats

stats.kstest(data_df["salary_standardized"], stats.norm.cdf, args=(data_df["salary_standardized"].mean(), data_df["salary_standardized"].std()))
```

```python
import seaborn as sns

sns.histplot(data_df["salary_standardized"],bins=30,kde=True)
```

```python
sns.boxplot(data = data_df["salary_standardized"])
```

```python
#STANDARD SALARY VS REMOTE

remote=data_df[data_df["work_from_home"]==True]
sns.histplot(data=remote,x="salary_standardized",bins=30,kde=True)
```

```python
stats.kstest(remote["salary_standardized"], stats.norm.cdf, args=(remote["salary_standardized"].mean(),remote["salary_standardized"].std()))
```

```python
#STANDARD SALARY VS NON-REMOTE
non_remote=data_df[data_df["work_from_home"]==False]
sns.histplot(data=non_remote,x="salary_standardized",bins=30,kde=True)
```

```python
stats.kstest(non_remote["salary_standardized"], stats.norm.cdf, args=(non_remote["salary_standardized"].mean(),non_remote["salary_standardized"].std()))
```

```python
sns.boxplot(data = data_df, x = "work_from_home", y = "salary_standardized")
```
